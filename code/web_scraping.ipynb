```python
import time
# from selenium.webdriver import Chrome # Use this if you are running locally.
from selenium.webdriver.common.by import By # This is for the By class, which is used to locate elements
from selenium.webdriver.common.keys import Keys # This is for the Keys class, which is used to send keyboard keys
from selenium.webdriver.support.ui import WebDriverWait # This is for the WebDriverWait class, which is used to wait for elements to be present
from selenium.webdriver.support import expected_conditions as EC # This is for the expected_conditions module, which is used to define conditions for waiting


video_url = 'https://www.youtube.com/watch?v=8Tniytx0ow0' # Video on neurodivergence and relationships

# with Chrome() as driver: # Use this if you are running locally.
with gs.Chrome() as driver:
    wait = WebDriverWait(driver,10) # This is the maximum time to wait for an element to be present
    driver.get(video_url)

    for item in range(10): # by increasing the highest range you can get more content
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END) # This is to scroll down the page and load more comments. The send_keys method is used to send the END key to scroll down the page. The visibility_of_element_located method is used to wait for the body tag to be visible before sending the key. The By.TAG_NAME is used to locate the body tag. The Keys.END is used to send the END key to scroll down the page.
        time.sleep(3)

    comments = []
    for comment in wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#comment #content-text"))): # This is to get all the comments. The presence_of_all_elements_located method is used to wait for all the comments to be present before getting them. The By.CSS_SELECTOR is used to locate the comments using CSS selectors. The #comment #content-text is used to locate the comments using CSS selectors.
        print(comment.text)
        comments.append(comment.text)

print(len(comments))
```

## Saving as a dataframe

```python
# Create dataframe from comments
import pandas as pd

df = pd.DataFrame(comments, columns=['comment'])
df.head()
```

```python
# Save dataframe to CSV
df.to_csv('comments.csv', index=False)
```
